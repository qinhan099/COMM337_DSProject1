### App Reviews
- What is the distribution of number of reviews in this data? Visualize it and discuss.
- Use scatter plot to show the relationship of ratings (as x axis) and reviews (as y axis) in this data, and interpret your results.

```python
df['Reviews'].describe()
```

```python
xReviews = df['Reviews']
plt.hist(xReviews, bins = 50, color = 'red', edgecolor='black')
plt.title('Densed Reviews Distribution')
plt.show()
```

```python
lower_limit = df['Reviews'].min()
upper_limit = df['Reviews'].max()
clipped_data = np.clip(df['Reviews'], lower_limit, upper_limit)

plt.hist(clipped_data, bins = 50, color = 'pink', edgecolor='black')
plt.title('Densed Reviews Distribution')
plt.show()
```

```python
## Scatter Plot

plt.scatter(df['Rating'], df['Reviews'], color='c')
plt.title('Relationship Rating/Reviews')
plt.xlabel('Ratings')
plt.ylabel('Reviews')
plt.grid()
plt.show()
```
